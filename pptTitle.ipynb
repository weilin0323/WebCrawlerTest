```python
import requests
from bs4 import BeautifulSoup 

r = requests.get("https://www.ptt.cc/bbs/MobileComm/index.html") #將此頁面的HTML GET下來
print(r.text) #印出HTML
```

```python
soup = BeautifulSoup(r.text,"html.parser") #將網頁資料以html.parser
sel = soup.select("div.title a") #取HTML標中的 <div class="title"></div> 中的<a>標籤存入sel
```

```python
print("練習來源:https://ithelp.ithome.com.tw/articles/10202121")
for s in sel:
    print(s["href"], s.text) 
```

```python
#爬取多頁

url = "https://www.ptt.cc/bbs/joke/index.html"
for i in range(3): #往上爬3頁
    r2 = requests.get(url)
    soup2 = BeautifulSoup(r2.text,"html.parser")
    sel2 = soup2.select("div.title a") #標題
    u = soup2.select("div.btn-group.btn-group-paging a") #a標籤
    print ("本頁的URL為"+url)
    url = "https://www.ptt.cc"+ u[1]["href"] #上一頁的網址

    for s2 in sel2: #印出網址跟標題
        print(s2["href"],s2.text)
    print()
```
